# TASK 1

### To find average of price1 on same week and create a csv file with two columns date,average price.

```python
sheet['Date and time'] = pd.to_datetime(sheet['date'])
sheet['Date']=sheet['Date and time'].dt.date
sheet['week'] = pd.factorize(sheet['Date and time'].dt.isocalendar().week)[0] + 1

sheet
```

```python
df = sheet.groupby(['Date','week'], as_index=False)['price1'].mean()
df
```

```python
df.rename(columns={'price1':'average price'}, inplace=True)
df1=df.drop('week',axis=1)
df1
```

# TASK 2

## Consider if  price1 > price2 ,that means ‘R’ and if price2 > price1 that means ‘G’,then we have to find 3 consecutive ‘R’.

```python
sheet2=sheet.drop(['date','Date and time'],axis=1)
sheet2
```

```python
sheet2['Code'] = sheet2.apply(lambda x: 'R' if x['price1'] >
                     x['price2'] else 'G', axis=1)
sheet2.head(20)
 
```

```python
g = sheet2['Code'].ne(sheet2['Code'].shift()).cumsum()
g = g[sheet2['Code'].eq('R')]
g.groupby(g).count().ge(3)
g.head(20)
```

```python
((sheet2['Code'] != 'R').cumsum().loc[sheet2['Code'] == 'R'].value_counts() == 3).head(30)
```
